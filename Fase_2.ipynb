### Estudo sobre a Importncia do Acompanhento da Sade Mental

Segundo a Mental Health America, mais de 44 milhes de adultos nos EUA possuem uma condio de sade mental. A sade mental de colaboradores na indstria de tecnologia  de grande preocupao devido ao tipo de ambiente competitivo tipicamente encontrado nestas empresas. Problemas de sade mental como ansiedade e depresso so prejudiciais ao bem-estar, estilo de vida, e produtividade das pessoas no local de trabalho. 

Anualmente, a OSMI conduz pesquisas que procuram mensurar as atitudes com relao a distrbios mentais nas empresas de tecnologia e examinar a frequncia de distrbios mentais entre funcionrios de tecnologia. Para este estudo de caso utilizaremos os resultados das pesquisas realizadas entre 2016 e 2020.
O estudo de caso utiliza dados coletados pela Open Sourcing Mental Illness (OSMI) (https://osmihelp.org/), que  uma ONG dedicada a conscientizar, educar e prover recursos para suportar pessoas com distrbios mentais nas empresas de tecnologia e comunidades open source.

#### 1-Existe uma diferena significativa de sade mental entre os gneros de colaboradores ao longo do tempo em que a pesquisa foi realizada?

```python
Q1 = new_df.groupby(['Genero', 'Ano'])['Ano'].count() / new_df.groupby(['Genero']) ['Ano'].count()
```

```python
# Visualizao em grfico
graficoQ1 = Q1.reset_index(name='Comparao')

# Plotando o grfico de barras agrupadas
plt.figure(figsize=(10, 6))
sns.barplot(x='Ano', y='Comparao', hue='Genero', data=graficoQ1)
plt.title('Comparao de Ocorrncias por Gnero e Ano')
plt.xlabel('Ano')
plt.ylabel('Comparao')
plt.legend(title='Gnero')
plt.show()
```

```python
# Visualizao tabela
print(Q1)
```

Comparando a incidncia na pesquisa, nota-se que em 2018 havia muito mais ocorrncia entre o gnero Masculino, essa porcentagem foi caindo ao passar dos anos e o gnero Feminino se destaca com maiores casos de ocorrncia por sade mental desde 2017 e nos anos posteriores. Em 2020 notamos uma queda de ocorrncia em ambos os gneros, mas devemos levar em considerao se a pesquisa foi feita em todo o ano de 2020 ou s nos primeiros meses, mas com os dados que temos podemos constatar a queda, porm a ocorrncia continua em maiores casos no gnero Feminino. 

#### 2-Existe uma diferena significativa de quem busca tratamento entre os gneros de colaboradores ao longo do tempo em que a pesquisa foi realizada?

```python
Q2 = new_df.groupby(['Genero', 'Ano', 'Tratamento_Profissional']) ['Ano'].count() / new_df.groupby(['Genero', 'Tratamento_Profissional']) ['Ano'].count()
```

```python
# Visualizao em grfico
graficoQ2 = Q2.reset_index(name='Comparao')

# Plotando grfico de barras facetado
g = sns.catplot(
    data=graficoQ2, kind="bar",
    x="Ano", y="Comparao", hue="Tratamento_Profissional",
    col="Genero", palette="viridis", height=4, aspect=0.7
)
g.fig.suptitle('Comparao de Ocorrncias por Gnero, Ano e Tratamento Profissional', y=1.02)
plt.show()
```

Com a visualizao dos dados podemos chegar a reposta de que h uma diferencia sim conforme se passam os anos, inicialmente h uma maior procura por ajuda profissional pelo gnero Masculino, porm h uma queda e a partir de 2017 h uma procura maior Feminina por tratamento profissional, isso ocorre at 2019 pois em 2020 podemos ver novamente um aumento de procura pelo gnero Masculino mesmo ocorrncia uma baixa de procura por ambos os sexos.
